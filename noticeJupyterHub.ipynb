# Notice JupyterHub :

Pour grer des fichiers sur le serveur JupyterHub du lyce, le plus pratique est d'utiliser la ligne de commande dans un terminal.

Pour ouvrir un terminal cliquez sur le bouton `Nouveau` en haut  droite de la page d'accueil puis choisissez `Terminal`.



Le texte `jupyter-pnom@r610b:~` indique que vous tes connect sur la machine nomme `r610b` sous le nom d'utilisateur `jupyter-pnom` et que le dossier de travail actuel est le dossier `~`, c'est--dire votre dossier personnel `/home/jupyter-pnom`.

Ce texte `jupyter-pnom@r610b:~$` s'appelle un prompt, une invite de commande, c'est--dire une invitation  entrer une commande systme... 

## Pour tlcharger un fichier :

```python
wget URL_de_tlchargement
```

Exemples :

- Pour tlcharger un notebook depuis le nbviewer jupyter, faire un clic droit sur la flche de tlchargement et choisir `Copier l'adresse du lien` puis la coller dans la ligne de commande telle que par exemple :

```python
wget https://nbviewer.jupyter.org/github/ericECmorlaix/1NSI_2020-2021/blob/master/index.ipynb
```

- Pour tlcharger une archive d'un dpot GitHub, faire un clic droit sur `Download ZIP`


et choisir `Copier l'adresse du lien` puis la coller dans la ligne de commande telle que par exemple :


```python
wget https://github.com/ericECmorlaix/1NSI_2020-2021/archive/master.zip
```
